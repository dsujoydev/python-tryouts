```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a synthetic dataset
X, y = make_classification(
    n_samples=1000,
    n_features=2,  # Reduced to 2 for scatter plot visualization
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Train a Naive Bayes classifier
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, y_pred_nb)

# Visualize Naive Bayes
def visualize_naive_bayes():
    # Scatter plot for predicted classes
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred_nb, cmap='coolwarm', alpha=0.7)
    plt.title("Naive Bayes Predicted Classes")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend(*scatter.legend_elements(), title="Classes")
    plt.show()

    # Confusion matrix heatmap
    cm_nb = confusion_matrix(y_test, y_pred_nb)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Blues')
    plt.title("Naive Bayes Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

visualize_naive_bayes()
```

```python
# Train an SVM classifier
svm_model = SVC(kernel='linear', random_state=42, probability=True)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, y_pred_svm)

# Visualize SVM
def visualize_svm():
    # Scatter plot for predicted classes
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred_svm, cmap='coolwarm', alpha=0.7)
    plt.title("SVM Predicted Classes")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend(*scatter.legend_elements(), title="Classes")
    plt.show()

    # Confusion matrix heatmap
    cm_svm = confusion_matrix(y_test, y_pred_svm)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Greens')
    plt.title("SVM Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()


visualize_svm()
```
